```python
filelist=glob('/home/dipak/Downloads/[1-9]*/*.html')
```

```python
filelist
```

```python
import json
tags = {}
with open('no_author.txt', 'w') as no_author, open('no_body.txt', 'w') as no_body:
    for file in filelist:
        html = open(file, 'r').read()
        article = Article('')
        article.set_html(html)
        article.parse()
        author = article.authors
        date = article.publish_date
        title = article.title
        body = article.text
        if(len(author) > 0):
            author1 = author[0]
        else:
            author1 = ''
            no_author.write(file+'\n')
        if (len(body) > 0):
            body1 = body
        else:
            body1 = ''
            no_body.write(file+'\n') 
        tags['date'] = str(date)
        tags['author'] = author1
        tags['body'] = body1
        tags['title'] = title
        tags['html_addr'] = file
        
        filetxt=file.split("/")[-1]
        f = open('/home/dipak/Iris/'+filetxt+'.txt', 'w')
        json.dump(tags, f)
        f.close()
            
```
